```python
#how many elemenst get fetched in loc and iloc when giving the range in native python and Pandas DF
lst = list(range(0,101))
print(' max:', max(lst)) # native python gives 0,...,100 elements
print(' len:', len(lst)) # means 101 elements
print()

df = pd.DataFrame(lst)
print('iloc:', len(df.iloc[0:100])) #pandas give 100 elements in iloc indexing, excluding the last one
print(' loc:', len(df.loc[0:100])) #pandas give 101 elements in loc indexing, including the last one
```

```python
#read the data from csv, skip rows and select only few columns in dataframes. setting 1 column as index
import numpy as np
import pandas as pd
olympic=pd.read_csv('olympic.csv')
olympic.head()
```

```python
#country who won maximum medal
olympic.groupby(['Country'])['Total'].sum().idxmax()
```

```python
#country appears how many times in data
olympic.Country.value_counts()
```

```python
#Remove special characters from Athlete names
olympic.Athlete=olympic.Athlete.str.decode('utf-8').str.encode('ascii', 'ignore')
```

```python
# Which is the best country having maximum medals, sort it by country
a=olympic.groupby(['Country']).Total.max()
best_ctry = a.sort_index()
best_ctry.head()
```

```python
#for each Athlete find the minimum and maximum year
import numpy as np
a_min=olympic.groupby(['Athlete']).Year.min()
a_max=olympic.groupby(['Athlete']).Year.max()
#wine_price_range=reviews.groupby('variety').price.agg([min, max]).sort_values(by=['min', 'max'], ascending=False)
pd.DataFrame(pd.concat([a_min,a_max],axis=1))
```

```python
#combination of countries and Athlete are most common
olympic['n']=0
df=olympic.groupby(['Country','Athlete']).n.count().sort_values(ascending=False)
print(df)
```

```python
# replace null values with 'Unknown' and give the count of each value in country
oly_ctry =olympic.Country.fillna('Unknown').value_counts().sort_values(ascending=False)
print(oly_ctry)
```
